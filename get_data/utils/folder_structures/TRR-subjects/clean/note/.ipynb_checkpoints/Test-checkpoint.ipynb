```python
import pandas as pd
import numpy as np
# Still does not correct for slightly off numbers like 1234 and 12340 being "different"
```

```python
df = pd.read_csv("../input/TRR-subjects_2004-2016_2016-09.csv.gz")
```

```python
cb_no = df['cb_no']
```

```python
def clean_cb(x):
    if isinstance(x, str):
        x = x.replace(".0", "").replace("0.", "").replace("-", "")
        if len(set(x)) == 1 or x == "12345678":
            return np.nan
        else:
            return int(x)
    elif np.isnan(x):
        return np.nan
    else:
        return int(x)
```

```python
v = pd.Series(cb_no).map(clean_cb)
```
